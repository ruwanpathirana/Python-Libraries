```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
randomNum=np.random.rand(10)
```

```python
print(randomNum)
```

```python
style.use('ggplot')  #select plot style
plt.plot(randomNum,'g',label='line one',linewidth=2) #plot random num
plt.xlabel('Range')  #x axis=no of random num(index)
plt.ylabel('Numbers')  #y axis=actual random num
plt.title('First Plot')
```

```python
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
team='Kandy','Jaffna','Galle','Dambulla','Colombo'
percentile1=[40,30,15,15,10]
```

```python
plt.figure(figsize=(10,10))
explode=(0.05,0,0,0,0)  #explode largest pie in data set
plt.pie(percentile1,labels=team,explode=explode,autopct='%1.1f%%',startangle=70)
plt.axis('equal')  #set axis equal to draw pie as circle
plt.title('LPL2022')
plt.show()
```
